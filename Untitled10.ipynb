```python
# Importing Seaborn
import seaborn as sns
%matplotlib inline

# Loading a dataset from seaborn
tips_dataset = sns.load_dataset('tips')
tips_dataset.head()
```

```python
# Use distplot
sns.distplot(tips_dataset['total_bill'], bins = 30, kde=False)
```

```python
# Estimating tip with respect to total bill
sns.relplot(x = 'total_bill', y ='tip', data = tips_dataset)
```

```python
sns.relplot(x = 'total_bill', y='tip', data = tips_dataset, hue = 'sex')
```

```python
sns.pairplot(tips_dataset, hue='sex', palette='coolwarm')
```

```python
tips_dataset.hist()
```

```python
sns.countplot(x='sex', data = tips_dataset)
```

```python
import pandas as pd
tips_dataset.groupby('sex').count()
```

```python
def compare(a, b):
    if a > b:
        print(a, ' > ', b)
    elif a < b:
        print(a, ' < ', b)
    else:
        print(a, ' = ', b)
compare(7,8)
x = 12
y = 5
compare(x, y)
compare(x, 12)
```
