```python
import matplotlib.pyplot as plt
import numpy as np

# Diasfalizei oti emfanizontai ola ta plot mas
%matplotlib inline 

# Creating plot values
x = np.linspace(0, 10) # x-axis/ time-scale
y = np.cos(x) # corresponding cosine values
plt.plot(x,y)
```

```python
# Mporoume na sxediasoyme polles sunarthseis se ena graph
z = np.sin(x)
plt.plot(x,y,x,z);plt.show()
```

```python
plt.plot(x,y,x,z)
plt.xlabel('Time axis') # Labeling x-axis
plt.ylabel('Magnitude') # Labeling y-axis
plt.title('Sine and Cosine waves') # graph title
plt.legend(['Cos','Sin'], loc=0) # Loc=0 means best location
plt.show() #printing
```

```python
# Parallhla Plots
plt.subplot(1,2,1) # plot 1
plt.plot(x,y)
plt.title('Cosine plot')
plt.subplot(1,2,2) # plot 2
plt.plot(x,z)
plt.title('Sine plot')
plt.tight_layout() # Entolh apofyghs epikalypshs twn plots
plt.show()
```

```python
# Parallhla Plots with colours and markers
plt.subplot(1,2,1) # plot 1
plt.plot(x,y, 'r-x') # red plot me -x marker
plt.title('Cosine plot')
plt.subplot(1,2,2) # plot 2
plt.plot(x,z, 'g-o') # green plot me -o marker
plt.title('Sine plot')
plt.tight_layout() # Entolh apofyghs epikalypshs twn plots
plt.show()
```

```python
# Aplo diagramma diasporas
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,2,1,4,5,2]

plt.scatter(x,y, label='Τυχαιες τιμες', color='r', s=25, marker='o')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Παραδειγμα με τυχαιες τιμες')
plt.legend()
plt.show()
```

```python
# Diagramma diasporas me omades
weight1=[57,58.2,58.6,59.6,59.8,60.2,60.5,60.6,60.7,61.3,61.3,61.4,61.8,61.9,62.3]
height1=[100.7,195.6,94.3,127.1,111.7,159.7,135,149.9,124.3,112.9,176.7,110.2,123.9,161.9,107.8]

weight2=[62.9,63,63.2,63.2,63.3,63.4,63.4,63.4,63.5,63.6,63.7,64.1,64.3,64.3,64.7,64.8,65]
height2=[151.8,156.3,136.1,124.2,156.3,130,181.2,255.9,163.1,123.1,119.5,179.9,114.8,174.1,108.8,105.4,141.4]

weight3=[69.2,69.2,69.4,69.7,70,70.3,70.8,71,71.1,71.7,71.9,72.4,73,73.1,76.2]
height3=[166.8,172.9,193.8,137.4,162.4,137.1,169.1,237.1,189.1,179.3,174.8,213.3,198,191.1,220.6]

weight=np.concatenate((weight1, weight2, weight3))
height=np.concatenate((height1, height2, height3))

color_array = ['b'] * 15 + ['g'] * 17 + ['r'] * 15
plt.scatter(weight, height, marker = '*', c = color_array)

plt.xlabel('weight', fontsize = 16)
plt.ylabel('height', fontsize = 16)
plt.title('grouped scatter plot - height vs weight', fontsize = 20)
plt.show()
```

```python
days = [1,2,3,4,5]

sleeping =[7,8,6,11,7]
eating   =[2,3,4,3,2]
working  =[7,8,7,2,2]
playing  =[8,5,7,8,13]

plt.plot([],[],color='m', label='Sleeping', linewidth=5)
plt.plot([],[],color='c', label='Eating', linewidth=5)
plt.plot([],[],color='r', label='Working', linewidth=5)
plt.plot([],[],color='k', label='Playing', linewidth=5)

plt.stackplot(days, sleeping,eating,working,playing, colors=['m','c','r','k'])

plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nCheck it out')
plt.legend()
plt.show()
```

```python
# Diagramma pitas, opou oi feres tha diataxthoun kai tha sxediasastoun
labels='Frogs','Hogs','Dogs','Logs'
sizes=[15,30,45,10]
explode=(0,0.1,0,0) # kanei explode mono sth 2h feta("Hogs")

fig1,ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
       shadow=True, startangle=90)
ax1.axis('equal') # H isotimh analogia diastasewn diasfalizei oti h pita sxediazetai ws kuklos

plt.show()
```
